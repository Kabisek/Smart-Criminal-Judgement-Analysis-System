```python
import pandas as pd
import numpy as np
import pickle

print("="*70)
print("✅ LOADING WEEK 6 DATA FOR WEEK 7 (FIXED)")
print("="*70)
print()

# Load the FULL feature matrices
print("Loading full feature matrices...")
X_train_full = pd.read_csv('X_train_bert.csv')
X_test_full = pd.read_csv('X_test_bert.csv')

print(f"✓ X_train (full): {X_train_full.shape}")
print(f"✓ X_test (full): {X_test_full.shape}")
print()

# Load target variables
print("Loading target variables...")
y_train = np.load('y_train_outcome.npy')
y_test = np.load('y_test_outcome.npy')

print(f"✓ y_train: {y_train.shape}")
print(f"✓ y_test: {y_test.shape}")
print()

# Load label encoder
print("Loading label encoder...")
with open('label_encoder_outcome.pkl', 'rb') as f:
    label_encoder = pickle.load(f)

class_names = label_encoder.classes_
print(f"✓ Classes: {list(class_names)}")
print()

# Load SELECTED feature names (the 150 features used in Week 6)
print("Loading selected features...")
with open('selected_features.pkl', 'rb') as f:
    selected_features = pickle.load(f)
print(f"✓ Selected features: {len(selected_features)} features")
print()

# ⭐ SELECT ONLY THE 150 FEATURES AND CONVERT TO NUMPY ARRAYS
print("Selecting the 150 features used in Week 6...")
X_train = X_train_full[selected_features].values  # ← Convert to numpy array
X_test = X_test_full[selected_features].values    # ← Convert to numpy array

print(f"✓ X_train (selected): {X_train.shape}")
print(f"✓ X_test (selected): {X_test.shape}")
print()

# Load previous model (XGBoost + SMOTE)
print("Loading Week 6 model...")
with open('appeal_outcome_model_xgboost_smote.pkl', 'rb') as f:
    previous_model = pickle.load(f)
print("✓ Previous model loaded")
print()

# Verify data
print("="*70)
print("📊 DATA VERIFICATION")
print("="*70)
print(f"Features (full): {X_train_full.shape[1]}")
print(f"Features (selected): {X_train.shape[1]}")
print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
print(f"Target classes: {len(class_names)}")
print()

print(f"Class distribution (train): {np.bincount(y_train)}")
print(f"  - Appeal_Allowed: {np.sum(y_train == 0)} ({np.sum(y_train == 0)/len(y_train)*100:.1f}%)")
print(f"  - Appeal_Dismissed: {np.sum(y_train == 1)} ({np.sum(y_train == 1)/len(y_train)*100:.1f}%)")
print(f"  - Partly_Allowed: {np.sum(y_train == 2)} ({np.sum(y_train == 2)/len(y_train)*100:.1f}%)")
print()

print(f"Class distribution (test): {np.bincount(y_test)}")
print(f"  - Appeal_Allowed: {np.sum(y_test == 0)} ({np.sum(y_test == 0)/len(y_test)*100:.1f}%)")
print(f"  - Appeal_Dismissed: {np.sum(y_test == 1)} ({np.sum(y_test == 1)/len(y_test)*100:.1f}%)")
print(f"  - Partly_Allowed: {np.sum(y_test == 2)} ({np.sum(y_test == 2)/len(y_test)*100:.1f}%)")
print()

# Quick baseline check
print("="*70)
print("📈 WEEK 6 BASELINE PERFORMANCE")
print("="*70)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred_baseline = previous_model.predict(X_test)
baseline_acc = accuracy_score(y_test, y_pred_baseline)

print(f"✓ Baseline Accuracy: {baseline_acc:.4f} ({baseline_acc*100:.2f}%)")
print()

print("Classification Report:")
print(classification_report(y_test, y_pred_baseline, target_names=class_names))
print()

print("Confusion Matrix:")
cm = confusion_matrix(y_test, y_pred_baseline)
print(cm)
print()

# Per-class accuracy
print("Per-class Accuracy:")
for i, class_name in enumerate(class_names):
    class_acc = cm[i, i] / cm[i].sum() if cm[i].sum() > 0 else 0
    print(f"  - {class_name}: {class_acc:.4f} ({class_acc*100:.1f}%)")
print()

print("="*70)
print("✅ ALL DATA LOADED SUCCESSFULLY!")
print("✅ READY FOR WEEK 7 ADVANCED MODELING!")
print("="*70)
print()

print("📌 SUMMARY:")
print(f"  • Week 6 Model: XGBoost + SMOTE")
print(f"  • Features: {X_train.shape[1]} (selected from {X_train_full.shape[1]})")
print(f"  • Baseline Accuracy: {baseline_acc*100:.2f}%")
print(f"  • Ready for hyperparameter tuning, ensembles, or deep learning!")
print("="*70)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import RandomizedSearchCV, StratifiedKFold
from sklearn.metrics import (accuracy_score, classification_report, 
                             confusion_matrix, f1_score, make_scorer)
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
import pickle
import time
from datetime import datetime

print("="*70)
print("🎯 WEEK 7: HYPERPARAMETER TUNING WITH RANDOMIZED SEARCH")
print("="*70)
print(f"Started: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print()

# ============================================================
# STEP 1: DEFINE HYPERPARAMETER SEARCH SPACE
# ============================================================

print("STEP 1: Defining hyperparameter search space...")
print("-"*70)

param_distributions = {
    # Learning rate: Lower = more conservative, better generalization
    'learning_rate': [0.01, 0.05, 0.1, 0.15, 0.2],
    
    # Number of boosting rounds
    'n_estimators': [100, 200, 300, 500, 700, 1000],
    
    # Maximum depth of trees: Controls model complexity
    'max_depth': [3, 4, 5, 6, 7, 8, 10],
    
    # Minimum child weight: Higher = more conservative
    'min_child_weight': [1, 3, 5, 7, 10],
    
    # Subsample: Fraction of samples used per tree
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],
    
    # Column sampling by tree
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],
    
    # L1 regularization (Lasso)
    'reg_alpha': [0, 0.01, 0.1, 0.5, 1, 5],
    
    # L2 regularization (Ridge)
    'reg_lambda': [0.01, 0.1, 0.5, 1, 5, 10],
    
    # Gamma: Minimum loss reduction to make split
    'gamma': [0, 0.1, 0.2, 0.5, 1],
    
    # Scale weight for minority class
    'scale_pos_weight': [1, 2, 3, 5]
}

print(f"✓ Total parameter combinations: {np.prod([len(v) for v in param_distributions.values()]):,}")
print(f"✓ We'll sample: 100 random combinations")
print()

# ============================================================
# STEP 2: APPLY SMOTE TO TRAINING DATA
# ============================================================

print("STEP 2: Applying SMOTE to handle class imbalance...")
print("-"*70)

# Calculate sampling strategy to balance classes
original_dist = np.bincount(y_train)
print(f"Original distribution: {original_dist}")

# Apply SMOTE
smote = SMOTE(random_state=42, k_neighbors=5)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

smote_dist = np.bincount(y_train_smote)
print(f"After SMOTE: {smote_dist}")
print(f"✓ Training samples: {len(y_train)} → {len(y_train_smote)}")
print()

# ============================================================
# STEP 3: SETUP RANDOMIZED SEARCH WITH CV
# ============================================================

print("STEP 3: Configuring RandomizedSearchCV...")
print("-"*70)

# Base model
base_model = XGBClassifier(
    objective='multi:softmax',
    num_class=3,
    random_state=42,
    eval_metric='mlogloss',
    use_label_encoder=False
)

# Custom scorer: macro F1 (treats all classes equally)
scorer = make_scorer(f1_score, average='macro')

# Stratified K-Fold for cross-validation
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=base_model,
    param_distributions=param_distributions,
    n_iter=100,  # Try 100 random combinations
    scoring=scorer,
    cv=cv_strategy,
    verbose=2,
    random_state=42,
    n_jobs=-1,  # Use all CPU cores
    return_train_score=True
)

print("✓ Scorer: Macro F1-Score (balances all classes)")
print("✓ CV Strategy: 5-Fold Stratified")
print("✓ Iterations: 100 random combinations")
print("✓ Parallel jobs: All CPU cores")
print()

# ============================================================
# STEP 4: FIT THE MODEL (THIS WILL TAKE TIME)
# ============================================================

print("="*70)
print("⏳ STEP 4: RUNNING HYPERPARAMETER SEARCH...")
print("="*70)
print("This may take 10-30 minutes depending on your machine...")
print("Grab a coffee! ☕")
print()

start_time = time.time()

# Fit the randomized search
random_search.fit(X_train_smote, y_train_smote)

end_time = time.time()
elapsed_time = end_time - start_time

print()
print("="*70)
print(f"✅ SEARCH COMPLETED in {elapsed_time/60:.1f} minutes")
print("="*70)
print()

# ============================================================
# STEP 5: ANALYZE RESULTS
# ============================================================

print("STEP 5: Best hyperparameters found...")
print("-"*70)

best_params = random_search.best_params_
best_cv_score = random_search.best_score_

print("Best parameters:")
for param, value in best_params.items():
    print(f"  • {param}: {value}")
print()

print(f"✓ Best CV F1-Score (macro): {best_cv_score:.4f}")
print()

# Get the best model
best_model = random_search.best_estimator_

# ============================================================
# STEP 6: EVALUATE ON TEST SET
# ============================================================

print("="*70)
print("📊 STEP 6: EVALUATION ON TEST SET")
print("="*70)
print()

# Predictions
y_pred_tuned = best_model.predict(X_test)
y_pred_baseline = previous_model.predict(X_test)

# Accuracy
acc_tuned = accuracy_score(y_test, y_pred_tuned)
acc_baseline = baseline_acc

# F1 Scores
f1_macro_tuned = f1_score(y_test, y_pred_tuned, average='macro')
f1_macro_baseline = f1_score(y_test, y_pred_baseline, average='macro')

f1_weighted_tuned = f1_score(y_test, y_pred_tuned, average='weighted')
f1_weighted_baseline = f1_score(y_test, y_pred_baseline, average='weighted')

print("PERFORMANCE COMPARISON:")
print("-"*70)
print(f"{'Metric':<25} {'Baseline':<15} {'Tuned':<15} {'Improvement':<15}")
print("-"*70)
print(f"{'Accuracy':<25} {acc_baseline:<15.4f} {acc_tuned:<15.4f} {(acc_tuned-acc_baseline)*100:+.2f}%")
print(f"{'F1-Score (macro)':<25} {f1_macro_baseline:<15.4f} {f1_macro_tuned:<15.4f} {(f1_macro_tuned-f1_macro_baseline)*100:+.2f}%")
print(f"{'F1-Score (weighted)':<25} {f1_weighted_baseline:<15.4f} {f1_weighted_tuned:<15.4f} {(f1_weighted_tuned-f1_weighted_baseline)*100:+.2f}%")
print()

# Classification Report
print("DETAILED CLASSIFICATION REPORT (TUNED MODEL):")
print("-"*70)
print(classification_report(y_test, y_pred_tuned, target_names=class_names))
print()

# Confusion Matrix
print("CONFUSION MATRIX (TUNED MODEL):")
print("-"*70)
cm_tuned = confusion_matrix(y_test, y_pred_tuned)
print(cm_tuned)
print()

# Per-class accuracy
print("PER-CLASS ACCURACY COMPARISON:")
print("-"*70)
cm_baseline = confusion_matrix(y_test, y_pred_baseline)

for i, class_name in enumerate(class_names):
    acc_base = cm_baseline[i, i] / cm_baseline[i].sum() if cm_baseline[i].sum() > 0 else 0
    acc_tune = cm_tuned[i, i] / cm_tuned[i].sum() if cm_tuned[i].sum() > 0 else 0
    improvement = (acc_tune - acc_base) * 100
    
    print(f"{class_name:<20} Baseline: {acc_base:.4f} | Tuned: {acc_tune:.4f} | {improvement:+.2f}%")
print()

# ============================================================
# STEP 7: SAVE THE TUNED MODEL
# ============================================================

print("="*70)
print("💾 STEP 7: SAVING TUNED MODEL")
print("="*70)

# Save best model
with open('appeal_outcome_model_tuned_week7.pkl', 'wb') as f:
    pickle.dump(best_model, f)
print("✓ Saved: appeal_outcome_model_tuned_week7.pkl")

# Save best parameters
with open('best_hyperparameters_week7.pkl', 'wb') as f:
    pickle.dump(best_params, f)
print("✓ Saved: best_hyperparameters_week7.pkl")

# Save results summary
results_summary = {
    'timestamp': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
    'search_time_minutes': elapsed_time / 60,
    'best_cv_score': best_cv_score,
    'best_params': best_params,
    'baseline_accuracy': acc_baseline,
    'tuned_accuracy': acc_tuned,
    'baseline_f1_macro': f1_macro_baseline,
    'tuned_f1_macro': f1_macro_tuned,
    'improvement_accuracy': (acc_tuned - acc_baseline) * 100,
    'improvement_f1_macro': (f1_macro_tuned - f1_macro_baseline) * 100
}

results_df = pd.DataFrame([results_summary])
results_df.to_csv('hyperparameter_tuning_results_week7.csv', index=False)
print("✓ Saved: hyperparameter_tuning_results_week7.csv")
print()

# ============================================================
# FINAL SUMMARY
# ============================================================

print("="*70)
print("🎉 WEEK 7 HYPERPARAMETER TUNING COMPLETE!")
print("="*70)
print()
print("📊 FINAL RESULTS:")
print(f"  • Baseline Accuracy: {acc_baseline*100:.2f}%")
print(f"  • Tuned Accuracy: {acc_tuned*100:.2f}%")
print(f"  • Improvement: {(acc_tuned-acc_baseline)*100:+.2f}%")
print()
print(f"  • Baseline F1 (macro): {f1_macro_baseline:.4f}")
print(f"  • Tuned F1 (macro): {f1_macro_tuned:.4f}")
print(f"  • Improvement: {(f1_macro_tuned-f1_macro_baseline)*100:+.2f}%")
print()
print("📁 FILES SAVED:")
print("  • appeal_outcome_model_tuned_week7.pkl")
print("  • best_hyperparameters_week7.pkl")
print("  • hyperparameter_tuning_results_week7.csv")
print()
print("="*70)
print("✅ READY FOR FURTHER IMPROVEMENTS OR DEPLOYMENT!")
print("="*70)
```
